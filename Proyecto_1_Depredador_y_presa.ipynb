```python
# CREANDO EL ESPACIO

#VARIABLES
# 0 -> Espacio vacios
# 1 -> Conejos
# 2 -> Lobos
# n -> Dimensiones de la matriz
n=8
tipos=([0,1,2])
espacio=[]
for _ in range(n):
  espacio.append(random.choices(tipos,weights=[0.5,2,1],k=8))

espacio=np.array(espacio)
print(espacio)
```
